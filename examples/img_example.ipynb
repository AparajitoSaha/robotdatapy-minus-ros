```python
import robotdatapy as rdp
from robotdatapy import data

import matplotlib.pyplot as plt
import cv2 as cv
```

## Bag file location and other data parameters

```python
bag_file_path = "/media/masonbp/T71/data/2024_07_highbay_overhead/exp1/runB.bag"
img_topic = "/overhead_cam/d455/color/image_raw/compressed"
compressed = True

# data params
time_tol = 0.5
```

## Load ImgData class

```python
img_data = rdp.data.ImgData.from_bag(
    path=bag_file_path,
    topic=img_topic,
    time_tol=time_tol,
    compressed=compressed
)
```

## Access camera image data

Use matplotlib to show an image

```python
img_data.show(img_data.t0)
```

Access and edit image

```python
img = img_data.img(img_data.t0 + 20)

# write "walking around the highbay"
cv.putText(img, "walking around the highbay", (10, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2)
plt.imshow(img[:, :, ::-1]) # convert BGR to RGB
```
